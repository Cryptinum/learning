```python
import warnings, requests
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np
from datetime import datetime
import ciso8601, time

# 全局设置
np.set_printoptions(suppress=True)
warnings.filterwarnings("ignore")  # Suppress warnings.
pd.options.display.max_rows = 10  # Display no more than 10 rows.
pd.set_option('display.float_format', lambda x: '%.2f' % x) # Round to 0.01.
```

```python
path = r"D:\test.csv"

df = pd.read_csv(path, index_col='Index')
```

```python
df.sort_index()
df.set_index(pd.Index(range(1,652)))
```

```python
grouped = df.groupby('Location')

pd.set_option('display.max_rows', 500)
counted = grouped.count()
counted['Name'] = counted['Period'] / counted['Period'].sum()
counted['Continent'] = grouped.nth(0)['Continent']

counted.style.format({'Name': "{:.2%}"}) # Style as percentage.
counted.columns = ['Numbers', 'Pct', 'Continent']

counted.sort_values(by='Numbers', ascending=False, inplace=True)

path = r"D:\aaa.csv"
counted.to_csv(path, encoding="utf-8-sig")

counted
```

```python
# 搜索关键词
path = r"D:\elite.csv"
conf = pd.read_csv(path, encoding="utf-8-sig")
print(conf.shape)

search = ["生态","环境","林业","水利","气象","自然","地理"]

#search = ["大数据"]

# conf['Title'].str.contains(search[0], case=False) 搜索了所有包含'search'字段的标题的序号，并用loc进行定位
if len(search) == 1:
    loced = conf.loc[conf['Title'].str.contains(search[0], case=False)]
else:
    #loced = conf.loc[conf.Title.apply(lambda sentence: all(word in sentence for word in search))] # 全部匹配
    loced = conf.loc[conf.Title.apply(lambda sentence: any(word in sentence for word in search))] # 任一匹配

print(loced.shape)
loced
```
